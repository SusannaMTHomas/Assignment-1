# Importing required libraries.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. Read the dataset to python environment.

```python
data=pd.read_excel('iris.xls')
```

# 2. Display the columns in the dataset.

```python
data.columns
```

# 3. Calculate the mean of each column of the dataset.

```python
data.mean()
```

# 4. Check for the null values present in the dataset.

```python
data.isnull().sum()
```

Inference: No values are missing in the dataset.

# 5. Perform meaningful visualizations using the dataset. Bring at least 3 visualizations.

# 5.1 Count of each species

```python
plt.figure(figsize=(8,8))
plt.title('Count of species')
sns.countplot(x='Classification',data=data)
plt.show()
```

Inference:<br> The data is well balanced as the data count of each species equals 50.

# 5.2 Comparison between various species based on sepal length and width

```python
plt.figure(figsize=(12,8))
plt.title('Comparison between various species based on sepal length and width')
sns.scatterplot(x='SL',y='SW',hue='Classification',data=data)
plt.show()
```

Inference: <br>Iris-setosa has smaller sepal length but higher width.<br> 
Iris-versicolor has average length as well as width.<br>
           Iris-virginica has larger sepal lengths and smaller sepal widths.

# 5.3 Comparison between various species based on petal length and width

```python
plt.figure(figsize=(12,8))
plt.title('Comparison between various species based on sepal length and width')
sns.scatterplot(x='PL',y='PW',hue='Classification',data=data)
plt.show()
```

Inference:<br>Iris-setosa has the smallest petal length and petal width.<br>
Iris-versicolor has average petal length and petal width.<br>
Iris-virginica has the highest petal length and petal width

# 5.4 Comparison between different attributes

```python
sns.pairplot(data,hue='Classification')
plt.show()
```

Inference:<br>There is a linear relationship between petal width and petal length.

# 5.5 Correlation between different attributes

```python
plt.figure(figsize=(10,8))
plt.title('Correlation Heatmap')
sns.heatmap(data.corr(),annot=True)
plt.show()
```

Inference: <br>
There is a close correlation between petal length and petal width as well as petal length and sepal length.

# 5.6 Distribution of various attributes

```python
plt.figure(figsize=(15,8))
plt.subplot(2,2,1)
sns.boxplot(x='Classification',y='SL',data=data)
plt.subplot(2,2,2)
sns.boxplot(x='Classification',y='PL',data=data)
plt.subplot(2,2,3)
sns.boxplot(x='Classification',y='SW',data=data)
plt.subplot(2,2,4)
sns.boxplot(x='Classification',y='PW',data=data)
plt.show()
```

Inference: <br>Statistics summary(median, quartiles, IQR, maximum and minimum) can be read from the above plots.<br>
Iris-setosa has smaller features and are less distributed.<br>
Iris-virginica has larger features and varies over a larger range. 

# 5.7 Violinplots

```python
plt.figure(figsize=(15,8))
plt.subplot(2,2,1)
sns.violinplot(x='Classification',y='SL',data=data)
plt.subplot(2,2,2)
sns.violinplot(x='Classification',y='PL',data=data)
plt.subplot(2,2,3)
sns.violinplot(x='Classification',y='SW',data=data)
plt.subplot(2,2,4)
sns.violinplot(x='Classification',y='PW',data=data)
plt.show()
```

Inference:<br>
The sepal width of all the three species is highly concentrated over the median.<br>
It is observed that Iris-setosa has all of its features closer to the median.
